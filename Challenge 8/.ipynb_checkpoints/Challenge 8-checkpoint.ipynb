## Problem 1: What are the top 3 and the bottom 3 states in terms of number of users?

```python
state_num = song.groupby('user_id').first()['user_state'].reset_index().groupby('user_state').count().reset_index()
state_num.columns = ['user_state','user_number']
state_num.sort_values(by = ['user_number'], ascending = False, inplace = True)
print(state_num.head(3))
print()
print(state_num.tail(3))
print()
print(state_num[state_num.user_number == 1])
```

As we can see, the top 3 states with most users are New York, California and Texas. The botoom 3 states are Nebraska, Arizona and Kansas. By the way, New Mexico, Idaho, North Dakota, Connecticut, Iowa, Rhode Island, Nebraska, Arizona and Kansas are all only have one user. They are all the bottom states.

## Problem 1: What are the top 3 and the bottom 3 states in terms of user engagement?

```python
song.head()
```

```python
user_song = song.groupby('user_id').count()['song_played']
state_song = song.groupby('user_id').first()['user_state'].reset_index()
state_song['num_of_songs_played'] = state_song.user_id.map(user_song)
state_song = state_song.groupby('user_state').mean()['num_of_songs_played'].reset_index()
state_song.sort_values(by = ['num_of_songs_played'], ascending = False, inplace = True)
print(state_song.head(3))
print()
print(state_song.tail(3))
```

From the table shown above, we can find that users in Nebraska, Alaska and Missisippi have highest average number of songs played. Users in Minnesota, Virginia and Kansas have lowest average number of songs played. 

## Problem 3: The CEO wants to send a gift to the first user who signed-up for each state. That is, the first user who signed-up from California, from Oregon, etc. Can you give him a list of those users?

```python
song['user_sign_up_date'] = pd.to_datetime(song.user_sign_up_date)
signup = song.groupby('user_id').min()['user_sign_up_date']
state_song = song.groupby('user_id').first()['user_state'].reset_index()
state_song['sign_up_date'] = state_song.user_id.map(signup)
first_sign = state_song.groupby('user_state').min()['sign_up_date'].reset_index()
result = pd.merge(state_song, first_sign, on = ['user_state','sign_up_date'], how = 'inner')
result
```

## Problem 4: Build a function that takes as an input any of the songs in the data and returns the most likely song to be listened next.

```python
song.head()
```

```python
s1 = song.sort_values(by = ['user_id','time_played'], ascending = [True,True]).reset_index(drop = True)
next_song = []
for i in range(s1.shape[0]):
    if i < s1.shape[0] - 1 and s1.loc[i,'user_id'] == s1.loc[i+1,'user_id']:
        next_song.append(s1.loc[i+1,'song_played'])
    else:
        next_song.append(np.nan)
s1['next_song'] = next_song
s1.head(10)
```

```python
s1.dropna(inplace = True)
s1.shape
```

```python
s1['time_played'] = pd.to_datetime(s1.time_played)
```

```python
s1['dow'] = s1.time_played.dt.dayofweek
s1['hour'] = s1.time_played.dt.hour
s1.head()
```
