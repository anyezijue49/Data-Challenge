```python
item = pd.read_csv('./item_to_id.csv')
purchase = pd.read_csv('./purchase_history.csv')
```

```python
df = purchase.copy()
df = pd.concat([df['user_id'],df['id'].str.get_dummies(sep=',')],1)
df.head()
```

```python
names = [df.columns[0]]
print(names)
for i in range(len(df.columns[1:])):
    j = df.columns[i+1]
    names.append(item[item.Item_id == int(j)].reset_index().loc[0,'Item_name'])
df.columns = names
```

```python
df['total'] = df.iloc[:,1:].sum(axis = 1)
df.head()
```

```python
purchase['total'] = purchase['id'].apply(lambda x: len(x.split(",")))
df['another'] = purchase['total']
df['difference'] = df['total'] - df['another']
sum(df['difference'])
```

```python
purchase.shape
```

```python
data = df.groupby('user_id').sum().reset_index()
data.head()
```

```python
data.drop(['total','another'], axis = 1, inplace = True)
```

```python
data['total'] = data.iloc[:,1:].sum(axis = 1)
data.head()
```

```python
data[data.total == data.total.max()][['user_id','total']]
```

## Problem 2:

```python
data.drop(['total','difference'], axis = 1, inplace = True)
```

```python
dict1 = {}
for i in data.columns[1:]:
    dict1[i] = data.loc[:,i].max()
```

```python
dict1.items()
```

```python
most_purchase = pd.DataFrame(list(dict1.items()), columns=['item', 'number'])
most_purchase.head()
```

```python
result = []
for j in range(len(most_purchase.item)):
    #print(j)
    i = most_purchase.item[j]
    #print(i)
    #print(most_purchase.iloc[j,1])
    result.append(list(data[data[i] == most_purchase.loc[j,'number']][["user_id",i]].user_id))
```

```python
most_purchase['user_id'] = result
most_purchase
```
